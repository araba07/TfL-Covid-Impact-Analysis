```python
tube_df = df[df['London Underground'] == 'Yes'].copy()
```

```python
tube_df.head(8)
```

```python
busiest_stations = tube_df.sort_values('En/Ex 2019', ascending=False)
```

```python
busiest_stations.head(1)
```

```python
lockdown_busiest_stations = tube_df.sort_values('En/Ex 2020', ascending=False)
```

```python
lockdown_busiest_stations.head(5)
```

```python
lockdown_quietest_stations = tube_df.sort_values('En/Ex 2020', ascending=True)
```

```python
lockdown_quietest_stations.head(5)
```

```python
tube_df['Pct_Drop'] = (tube_df['En/Ex 2020'] - tube_df['En/Ex 2019']) / tube_df['En/Ex 2019'] * 100
```

```python
tube_df.head(3)
```

```python
tube_df.sort_values('Pct_Drop', ascending=True)
```

```python
tube_df.groupby('LINES')['Pct_Drop'].mean().sort_values(ascending=True)
```

```python
zone_lookup = pd.read_csv('https://raw.githubusercontent.com/cusp-london/Spatial-Data-Analysis/master/Stations%202018.csv')
```

```python
zone_lookup[['NAME','Zone']].head()
```

```python
merged_df = pd.merge(tube_df, zone_lookup, left_on='Station', right_on='NAME')
```

```python
merged_df.info()
```

```python
merged_df.groupby('Zone')['Pct_Drop'].mean().sort_values(ascending=True).plot(kind='barh')
```
